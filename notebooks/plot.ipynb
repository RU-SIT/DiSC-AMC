```python
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
```

```python
# Set default template for a clean, paper-ready look
pio.templates.default = "plotly_white"

# Data from your table
data = {
    '# Bins': [5, 10, 20, 30],
    'Accuracy (%)': [45.50, 39.00, 38.00, 37.00]
}
df = pd.DataFrame(data)

# Define the style based on your example code (using 'PN Train' style)
main_color = 'rgba(31, 119, 180, 1)'  # Blue
bg_color = 'rgba(31, 119, 180, 0.3)' # Lighter, semi-transparent blue
marker_symbol = 'circle'

# Create the figure object
fig = go.Figure()

# Add the background trace (thick, semi-transparent line)
fig.add_trace(go.Scatter(
    x=df["# Bins"],
    y=df["Accuracy (%)"],
    mode='lines',
    line=dict(color=bg_color, width=12),
    hoverinfo='none' # Don't show hover info for the background line
))

# Add the foreground trace (thin line with markers)
fig.add_trace(go.Scatter(
    x=df["# Bins"],
    y=df["Accuracy (%)"],
    mode='lines+markers',
    name='Accuracy',
    line=dict(color=main_color, width=2),
    marker=dict(
        color=main_color,
        symbol=marker_symbol,
        size=22
    )
))

# Apply the detailed layout and styling from your example
fig.update_layout(
    title=dict(
        text='<b>Effect of Number of Bins on Accuracy</b>',
        x=0.5,
        font=dict(size=32)
    ),
    xaxis_title='<b># Bins</b>',
    yaxis_title='<b>Accuracy (%)</b>',
    font=dict(size=28, family="Arial, sans-serif"),
    showlegend=False,
    margin=dict(l=80, r=40, t=80, b=80),
    paper_bgcolor="white",
    plot_bgcolor="white",
)

# Style the X and Y axes
fig.update_xaxes(
    showline=True,
    linewidth=2,
    linecolor='black',
    mirror=True,
    ticks="outside",
    tickwidth=2,
    tickfont=dict(size=26, weight='bold')
)
fig.update_yaxes(
    showline=True,
    linewidth=2,
    linecolor='black',
    mirror=True,
    ticks="outside",
    tickwidth=2,
    tickfont=dict(size=26, weight='bold')
)

# Save the figure as a static image
fig.write_image("styled_bins_vs_accuracy.png", scale=3)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots

# Set default template for a clean, paper-ready look
pio.templates.default = "plotly_white"

# Data from your table, focusing on the consistent "# Bins = 5" experiment
data = {
    'top-k': [2, 3, 4, 5, 6, 10],
    '# Tokens': ['0.8K', '1.0K', '1.1K', '1.3K', '1.4K', '2.9K'],
    'Accuracy (%)': [49.50, 42.50, 44.50, 45.50, 44.50, 29.50]
}
df = pd.DataFrame(data)

# Convert '# Tokens' from string (e.g., '0.9K') to numeric values
df['Token Count'] = df['# Tokens'].str.replace('K', '').astype(float) * 1000

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Define styles for the two lines
style1 = {
    'color': 'rgba(31, 119, 180, 1)',  # Blue
    'bg_color': 'rgba(31, 119, 180, 0.3)',
    'symbol': 'circle'
}
style2 = {
    'color': 'rgba(255, 127, 14, 1)',  # Orange
    'bg_color': 'rgba(255, 127, 14, 0.3)',
    'symbol': 'star'
}

# --- Accuracy Trace (Left Y-axis) ---
# Background trace
fig.add_trace(go.Scatter(
    x=df["top-k"],
    y=df["Accuracy (%)"],
    mode='lines',
    line=dict(color=style1['bg_color'], width=12),
    hoverinfo='none',
    showlegend=False  # Hide from legend
), secondary_y=False)

# Foreground trace
fig.add_trace(go.Scatter(
    x=df["top-k"],
    y=df["Accuracy (%)"],
    mode='lines+markers',
    name='Accuracy',
    line=dict(color=style1['color'], width=2),
    marker=dict(
        color=style1['color'],
        symbol=style1['symbol'],
        size=22
    )
), secondary_y=False)


# --- Token Size Trace (Right Y-axis) ---
# Background trace
fig.add_trace(go.Scatter(
    x=df["top-k"],
    y=df["Token Count"],
    mode='lines',
    line=dict(color=style2['bg_color'], width=12),
    hoverinfo='none',
    showlegend=False # Hide from legend
), secondary_y=True)

# Foreground trace
fig.add_trace(go.Scatter(
    x=df["top-k"],
    y=df["Token Count"],
    mode='lines+markers',
    name='Token Size',
    line=dict(color=style2['color'], width=2),
    marker=dict(
        color=style2['color'],
        symbol=style2['symbol'],
        size=22
    )
), secondary_y=True)


# --- Layout and Styling ---
fig.update_layout(
    title=dict(
        text='<b>Effect of top-k Pruning on Accuracy and Token Size</b>',
        x=0.5,
        font=dict(size=32)
    ),
    xaxis_title='<b>top-k</b>',
    font=dict(size=28, family="Arial, sans-serif"),
    showlegend=True, # Show legend
    legend=dict(
        orientation="h",
        yanchor="top",
        y=-0.2,
        xanchor="right",
        x=1
    ),
    margin=dict(l=80, r=80, t=80, b=80),
    paper_bgcolor="white",
    plot_bgcolor="white",
)

# Style X and Y axes
fig.update_xaxes(
    showline=True,
    linewidth=2,
    linecolor='black',
    mirror=True,
    ticks="outside",
    tickwidth=2,
    tickfont=dict(weight='bold')
)
fig.update_yaxes(
    showline=True,
    linewidth=2,
    linecolor='black',
    mirror=True,
    ticks="outside",
    tickwidth=2,
    tickfont=dict(weight='bold')
)

# Set y-axis titles
fig.update_yaxes(title_text="<b>Accuracy (%)</b>", secondary_y=False, color=style1['color'])
fig.update_yaxes(title_text="<b>Number of Tokens</b>", secondary_y=True, color=style2['color'])


# Save and show the figure
fig.write_image("top-k_vs_accuracy_and_tokens.png", scale=5)
fig.show()
```
